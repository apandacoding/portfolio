```python
def most_used_word(text):
    # Default excluded words (generic nouns and conjunctions)

    excluded_words = ["the", "a", "an", "and", "or", "but", "for", "nor", "so", "yet", "to", "in", "on", "of", "with"]

    # Tokenize the text into words (assuming words are separated by spaces)
    words = re.findall(r'\b\w+\b', text.lower())

    # Remove excluded words
    words = [word for word in words if word not in excluded_words]

    # Count the occurrences of each word
    word_counts = Counter(words)

    # Find the most common word
    most_common_word, frequency = word_counts.most_common(1)[0]

    return most_common_word, frequency
```

```python
def description_filter(description):
  # Split text into sentences based on sentence-ending punctuation marks
  sentence_endings = re.compile(r'[.!?]')
  sentences = sentence_endings.split(description)

  # Remove empty sentences
  sentences = [s.strip() for s in sentences if s.strip()]
  length = len(sentences)

  sentences_to_keep = 3

  position = 0
  for i in range(sentences_to_keep):
    position = description.find('.', position + 1)



  # Extract the text up to the end of the third sentence
  extracted_text = description[0:position+1]
  return extracted_text
```

```python
import re

def extract_salary(text):
    # Regular expression pattern to match anything with a dollar sign
    pattern = r'\$\d+(?:,\d+)*'

    # Find all matches of the pattern in the text
    matches = re.findall(pattern, text)

    # Extract salary amounts from matches
    salaries = []
    for match in matches:
        # Remove the leading dollar sign and comma separators
        salary = int(match.lstrip('$').replace(',', ''))
        salaries.append(salary)

    salaries = [salary for salary in salaries if salary >= 10000]
    return salaries
```
